```python
#--import libraries

import pandas as pd
import numpy as np
from tqdm import tqdm

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
#---initialize stemmer,lemmatizer
stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()
```

```python
#--data reading
data_path="/content/gdrive/MyDrive/Learning_AI/NLP/dataset/kaggle_fake_news/train.csv"
df=pd.read_csv(data_path)
df.head()
```

```python
#--remove nan values
print(df.shape)
df=df.dropna()
print(df.shape)
```

```python
#--perform text preprocessing for a given column within dataframe

#--remove unwatend letters
#--convert to lower case
#--remove stop words

def df_text_preprocessing(df,col_list):

  '''
  df: input dataframe
  col_list: list of columns on which preprocessing needed
  df: output dataframe with modified text
  '''

  df[col_list]=df[col_list].apply(lambda x:x.str.lower())

  for col in col_list:
    df[col].replace('[^a-zA-Z]', ' ',regex=True,inplace=True)

    sent_list=list(df[col])
    #--remove stop words

    for i in tqdm(range(len(sent_list))):
      words=sent_list[i].split()
      words=[lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')]
      review=' '.join(words)
      sent_list[i]=review

    df[col]=sent_list
  return df
```

```python
col_list=['text']
filtered_df=df_text_preprocessing(df,col_list)
filtered_df.head(3)
```
