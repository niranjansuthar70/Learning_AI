```python
#--import required libraries
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
#--read dataset
data_path="/content/gdrive/MyDrive/Learning_AI/pytorch/dataset/diabetes.csv"
data=pd.read_csv(data_path)
data.head(2)
```

```python
X=data.drop('Outcome',axis=1).values
y=data['Outcome'].values
```

```python
#--divide data into train and test split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
X_train.shape
```

```python
X_train=torch.FloatTensor(X_train)
X_test=torch.FloatTensor(X_test)
y_train=torch.LongTensor(y_train)
y_train=torch.LongTensor(y_train)
```

```python
#--create model class

class ann_model(nn.Module):

  def __init__(self,input_features=8,hidden_layer1=10,hidden_layer2=10,output_features=2):

    super().__init__()

    self.fc1=nn.Linear(input_features,hidden_layer1)
    self.fc2=nn.Linear(hidden_layer1,hidden_layer2)
    self.output=nn.Linear(hidden_layer2,output_features)

  def forward(self,x):

    x=F.relu(self.fc1(x))
    x=F.relu(self.fc2(x))
    x=F.relu(self.output(x))

    return x
```

```python
torch.manual_seed(20)
```

```python
ann_model_trainer=ann_model()
ann_model_trainer.parameters
```

```python
loss_fn=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(ann_model_trainer.parameters(),lr=0.01)
optimizer
```

```python
epoch=10

combined_loss=[]

for i in range(epoch):

  y_pred=ann_model_trainer.forward(X_train)
  loss=loss_fn(y_pred,y_train)

  combined_loss.append(loss)

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```
