```python
import torch

print(torch.__version__)
```

```python
import numpy as np
```

```python
l=[1,2,3,4,5]

#--convert list to array
arr=np.array(l)
arr
```

```python
#--convert numpy array to tensor
t1=torch.from_numpy(arr)
t2=torch.tensor(arr)

print(t1)
print(t2)
```

```python
t1[0]=2
```

```python
print(arr)
print(t1)
print(t2)
```

```python
a=torch.tensor([1,2,3,4],dtype=torch.int32)
b=torch.tensor([1,1,1,1],dtype=torch.int32)
```

```python
c=torch.add(a,b)
c
```

```python
d=torch.matmul(a,b)
d
```

```python
a=torch.ones((3,2,3),dtype=torch.int32)
a
```
